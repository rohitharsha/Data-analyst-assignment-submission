```python
# 1)Convert minutes to human readable (example function)
def human_readable_minutes(minutes):
    hours = minutes // 60
    mins = minutes % 60
    if hours > 0:
        return f"{hours} hr{'s' if hours>1 else ''} {mins} minute{'s' if mins!=1 else ''}"
    else:
        return f"{mins} minute{'s' if mins!=1 else ''}"

# Examples
print(human_readable_minutes(130))
print(human_readable_minutes(110))
```

```python
# 2)Remove duplicates from string using loop (preserve order)
def unique_string(s):
    seen = set()
    out = []
    for ch in s:
        if ch not in seen:
            seen.add(ch)
            out.append(ch)
    return ''.join(out)

# Example
print(unique_string("banana"))  # "ban"
```
